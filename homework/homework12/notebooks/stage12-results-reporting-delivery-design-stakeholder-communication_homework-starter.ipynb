## Load Your Data
You can load your real results or use a generated synthetic dataset as fallback.

```python
data_path = Path('../data/final_results.csv')

if data_path.exists():
    df = pd.read_csv(data_path)
else:
    # Synthetic fallback dataset
    df = pd.DataFrame({
        'scenario': ['baseline','alt_impute','alt_outlier'],
        'return': [0.12, 0.11, 0.135],
        'volatility': [0.18, 0.185, 0.19],
        'sharpe': [0.56, 0.49, 0.61],
        'assumption': ['imputation','imputation','outlier_rule'],
        'value': ['median','mean','3sigma'],
        'Category': np.random.choice(['X','Y','Z'], 3),
        'MetricA': np.random.normal(75, 15, 3),
        'MetricB': np.random.normal(150, 30, 3),
        'Date': pd.date_range('2025-02-01', periods=3)
    })
df.head()
```

## Helper: Export Directory

```python
img_dir = Path('../deliverables/images')
img_dir.mkdir(parents=True, exist_ok=True)

def savefig(name):
    plt.tight_layout()
    plt.savefig(img_dir / name, dpi=300)
    print(f'Saved {name}')
```

## Chart 1: Riskâ€“Return Scatter

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='volatility', y='return', hue='scenario', s=80)
plt.title('Riskâ€“Return by Scenario')
plt.xlabel('Volatility')
plt.ylabel('Return')
savefig('risk_return.png')
plt.show()
```

Chart 1 Riskâ€“Return:
 - The baseline scenario shows moderate return and volatility, while the outlier-adjusted scenario achieves the highest return but also the highest volatility and the impute scenario has lowest return but high volatility as well.
 - Imputation (mean vs. median) slightly reduces return and increases volatility, suggesting sensitivity to how missing data is handled.
 - Trade-offs between risk and reward are visually clear, supporting scenario comparison.

## Chart 2: Return by Scenario (Bar Chart)

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='scenario', y='return')
plt.title('Return by Scenario')
savefig('return_by_scenario.png')
plt.show()
```

Chart 2 Return by Scenario:
 - Alt_outlier scenario delivers the highest return, while alt_impute scenario shows the lowest return and baseline scenario yields a moderate return.
 - Return is sensitive to assumption choices.
 - Imputation method (mean vs. median) and outlier treatment can materially shift performance.

## Chart 3: MetricA Over Time (Line Chart)

```python
plt.figure(figsize=(7,5))
sns.lineplot(data=df, x='Date', y='MetricA', hue='Category', marker='o')
plt.title('MetricA Over Time by Category')
plt.xlabel('Date')
plt.ylabel('MetricA')
savefig('metricA_over_time.png')
plt.show()
```

## Interpretations / Takeaways
- **Chart 1 takeaway:** 
 - The baseline scenario shows moderate return with moderate volatility.  
 - The outlier-adjusted scenario achieves the highest return but also the highest volatility, showing the trade-off between opportunity and risk.  
 - Assumption: Data cleaning choices (e.g., whether to remove outliers) directly affect the perceived returnâ€“risk balance.  
- Sensitivity: Small changes in data treatment can shift results significantly, meaning decisions are sensitive to methodological choices.  
- **Chart 2 takeaway:** 
 - Average returns differ by scenario, with baseline returns in the middle range.  
 - Removing outliers increases returns but may exaggerate upside potential, while alternative imputation methods can reduce returns.  
 - Assumption: Scenario definitions (baseline vs. adjusted) represent different but reasonable methodological paths.  
 - Sensitivity: Stakeholder decisions based on expected returns should account for the fact that outcomes vary by data treatment.  

- **Chart 3 takeaway:** 
 - Most categories show stable or slightly rising MetricA trends over time.  
 - A few categories display noticeable volatility, highlighting areas that may drive uncertainty in overall results.  
 - Assumption: Past observed trends in MetricA are representative of future performance, though this may not always hold.  
 - Sensitivity: If volatile categories deviate further, they could alter aggregate outcomes, making monitoring important.
